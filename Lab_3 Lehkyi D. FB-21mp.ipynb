```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
%matplotlib inline
```

# 1. Повнозв'язані нейронні мережі (Бінарна класифікація)

```python
#https://www.kaggle.com/datasets/mssmartypants/water-quality
water = pd.read_csv('waterQuality.csv')
```

```python
water = water.drop(labels=[7551,7568,7890],axis=0)
water['ammonia'] = water['ammonia'].astype(float)
water['is_safe'] = water['is_safe'].astype(int)
```

```python
water_features = water.drop(['is_safe'],axis=1) #X
water_target = water['is_safe'] #y


X_train, X_test, y_train, y_test = train_test_split(water_features, water_target, test_size=0.20)
```

```python
water_model_2_layers = keras.Sequential([
    layers.Dense(128, activation="relu"),
    layers.Dense(128, activation="relu"),
    layers.Dense(1, activation="sigmoid")
])

water_model_3_layers = keras.Sequential([
    layers.Dense(128, activation="relu"),
    layers.Dense(64, activation="relu"),
    layers.Dense(64, activation="relu"),
    layers.Dense(1, activation="sigmoid")
])

water_model_4_layers = keras.Sequential([
    layers.Dense(128, activation="relu"),
    layers.Dense(64, activation="relu"),
    layers.Dense(64, activation="relu"),
    layers.Dense(64, activation="relu"),
    layers.Dense(1, activation="sigmoid")
])
```

```python
water_model_2_layers.compile(optimizer="rmsprop",
              loss="binary_crossentropy",
              metrics=["accuracy"])

water_model_3_layers.compile(optimizer="rmsprop",
              loss="binary_crossentropy",
              metrics=["accuracy"])

water_model_4_layers.compile(optimizer="rmsprop",
              loss="binary_crossentropy",
              metrics=["accuracy"])
```

```python
history_2l = water_model_2_layers.fit(X_train,
                    y_train,
                    epochs=40,
                    validation_split=0.2)
```

```python
history_3l = water_model_3_layers.fit(X_train,
                    y_train,
                    epochs=35,
                    validation_split=0.2)
```

```python
history_4l = water_model_4_layers.fit(X_train,
                    y_train,
                    epochs=30,
                    validation_split=0.2)
```

```python
water_model_2_layers.summary()
water_model_3_layers.summary()
water_model_4_layers.summary()
```

```python
history_2l_dict = history_2l.history
history_2l_dict.keys()

history_3l_dict = history_3l.history

history_4l_dict = history_4l.history
```

```python
def build_graphics(history_dict_data,model_name):
    fig = plt.figure(figsize=(12,6))
    fig.add_subplot(121)
    loss_values = history_dict_data["loss"]
    val_loss_values = history_dict_data["val_loss"]
    epochs = range(1, len(loss_values) + 1)
    plt.plot(epochs, loss_values, "r", label="Training loss")
    plt.plot(epochs, val_loss_values, "g", label="Validation loss")
    plt.title(model_name+" loss")
    plt.xlabel("Epochs num")
    plt.ylabel("Loss")
    plt.legend()
    
    fig.add_subplot(122)
    acc_values = history_dict_data["accuracy"]
    val_acc_values = history_dict_data["val_accuracy"]
    epochs = range(1, len(acc_values) + 1)
    plt.plot(epochs, acc_values, "r", label="Training accuracy")
    plt.plot(epochs, val_acc_values, "g", label="Validation accuracy")
    plt.title(model_name+" accuracy")
    plt.xlabel("Epochs num")
    plt.ylabel("Accuracy")
    plt.legend()
    
    plt.show()
    
```

```python
build_graphics(history_2l_dict,"2 layers model")

build_graphics(history_3l_dict,"3 layers model")

build_graphics(history_4l_dict,"4 layers model")
```

```python
result = water_model_2_layers.evaluate(X_test,y_test)
print('Test score:', result[0]) 
print('Test accuracy:', result[1])

result = water_model_3_layers.evaluate(X_test,y_test)
print('Test score:', result[0]) 
print('Test accuracy:', result[1])

result = water_model_4_layers.evaluate(X_test,y_test)
print('Test score:', result[0]) 
print('Test accuracy:', result[1])
```

# 2. Згорткові нейронні мережі (Класифікація зображень)

Посилання на датасет: https://www.kaggle.com/datasets/utkarshsaxenadn/animal-image-classification-dataset

```python
import os
from PIL import Image
from sklearn.utils import shuffle
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

from tensorflow.keras.utils import load_img
from tensorflow.keras.utils import img_to_array
from tensorflow.keras.utils import array_to_img
from sklearn.model_selection import train_test_split
```

```python
classes = ['Dog','Cat','Tiger','Zebra']
```

```python
images_arr = []
images_names = []
classes_arr = []
folders = classes
class_id = 0

for folder in folders:
    for imgname in os.listdir(folder):
        image_sample = load_img(os.path.join(folder,imgname),target_size=(200,200))
        image_sample = img_to_array(image_sample)
        images_arr.append(image_sample)
        classes_arr.append(class_id)
        images_names.append(imgname)
    class_id+=1
            
images_arr = np.array(images_arr)
images_arr.shape
images_arr /= 255
images_arr[0]            
```

```python
img_data = {
    "array": images_names,
    "class": classes_arr
}
img_data = pd.DataFrame(img_data)
img_data
```

```python
sns.countplot(x='class', data = img_data)
```

### 0 - Dog
### 1 - Cat
### 2 - Tiger
### 3 - Zebra

```python
fig = plt.figure(figsize=(12,40))
for i in range(len(classes)):
    ax = fig.add_subplot(1,4,i+1)
    ax.set_title(classes[i])
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    image = array_to_img(images_arr[2000*i])
    plt.imshow(image)

    
    
```

```python
X_train, X_test, y_train, y_test = train_test_split(images_arr, img_data['class'], test_size=0.20)
y_train = np.asarray(y_train).astype(np.int32)
y_test = np.asarray(y_test).astype(np.int32)

y_train = tf.one_hot(y_train, 4)
y_test = tf.one_hot(y_test, 4)
```

```python
#Creating model

alpha = 0.02

model = keras.Sequential()

model.add(layers.Conv2D(filters=16, kernel_size=(3, 3), padding="same", input_shape=(200,200,3)))
model.add(layers.LeakyReLU(alpha=alpha))
model.add(layers.MaxPooling2D(pool_size=(2, 2), padding='same'))

model.add(layers.Conv2D(filters=32, kernel_size=(3, 3), padding="same"))
model.add(layers.LeakyReLU(alpha=alpha))
model.add(layers.MaxPooling2D(pool_size=(2, 2), padding='same'))

model.add(layers.Conv2D(filters=64, kernel_size=(3, 3), padding="same"))
model.add(layers.MaxPooling2D(pool_size=(2, 2), padding='same'))


model.add(layers.Flatten())

model.add(layers.Dense(64))
model.add(layers.LeakyReLU(alpha=alpha))

model.add(layers.Dense(4, activation="softmax"))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

model.summary()
```

```python
history_img = model.fit(X_train,
                    y_train,
                    epochs=13,
                    validation_split=0.2,
                    verbose = 1)
```

```python
history_img_dict = history_img.history

build_graphics(history_img_dict,"image classification model")
```

```python
model.evaluate(X_test,y_test)
```

# 3. Рекурентні нейронні мережі (Класифікація текстів)

Посилання на датасет:
https://www.kaggle.com/datasets/chandramoulinaidu/spam-classification-for-basic-nlp

```python
import re
import string
import nltk

from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from bs4 import BeautifulSoup
```

```python
spam_dataset = pd.read_csv('text_spam.csv')
spam_dataset = spam_dataset.drop(['FILE_NAME'], axis = 1)
spam_dataset.shape
```

```python
spam_dataset.head(10)
```

```python
spam_dataset['CATEGORY'].value_counts()
```

1 - Spam, 0 - Not spam

```python
stop_words = stopwords.words("english")
def clean_text(text):
    #видалення html тегів з тексту
    soup = BeautifulSoup(text)
    text = soup.get_text()
    text = text.lower()
    text = re.sub("@\S+", " ", text)
    text = re.sub("http*\S+", " ", text)
    text = re.sub("#\S+", " ", text)
    text = re.sub("\d", " ", text)
    text = re.sub('[%s]' % re.escape(string.punctuation), ' ', text)
    text = re.sub('\-\s\r\n\s{1,}|\-\s\r\n|\r\n', '', text)
    text = re.sub('\n', ' ', text)
    text = re.sub('\s',' ', text)
    text = ' '.join([word for word in text.split(' ') if word not in stop_words])
    #text = re.sub('\s[a-z]\s', " ", text)
    text = re.sub('\\b\\w{1,2}\\b', " ", text)
    odd_words = ['text','cdc','color','encoding','ffont','content','nbsp','2ffont','3darial','type','cfont','font','size','darial','list','charset','quoted','iso']
    for word in odd_words:
        text = re.sub('\s'+word+'\s', " ", text)
    text = re.sub('\s{2,}',' ', text)    
    return text
```

```python
spam_dataset['MESSAGE'] = spam_dataset['MESSAGE'].apply(clean_text)
spam_dataset.head(10)
```

```python
from keras.preprocessing.text import Tokenizer
```

```python
num_of_words = 5000
tokenizer = Tokenizer(num_of_words)
tokenizer.fit_on_texts(spam_dataset['MESSAGE'])
```

```python
textSequences = tokenizer.texts_to_sequences(spam_dataset['MESSAGE'])
text_data = tf.keras.preprocessing.sequence.pad_sequences(textSequences, maxlen=500)
text_target = spam_dataset['CATEGORY']
```

```python
model = keras.Sequential()
model.add(layers.Embedding(input_dim=num_of_words, output_dim=64,input_length=500))
model.add(layers.LSTM(units=128,return_sequences=True))
model.add(layers.LSTM(units=64))
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer="adam",
              loss="binary_crossentropy",
              metrics=["accuracy"])


```

```python
model.summary()
```

```python
X_train, X_test, y_train, y_test = train_test_split(text_data, text_target, test_size=0.20)
```

```python

history = model.fit(X_train,
                    y_train,
                    epochs=5,
                    batch_size=128,
                    validation_split=0.2,
                    shuffle = True,
                    verbose = 1)
```

```python
history_dict_t = history.history
build_graphics(history_dict_t,"text model")
```

```python
model.evaluate(X_test,y_test)
```
